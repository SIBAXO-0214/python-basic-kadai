```python
class Human:
    def __init__(self, name, age):
     self.name = name
     self.age =age

    def check_adult(self):
     if self.age >= 20:
      print(f"{self.name}は大人です。")
     else:
      print(f"{self.name}は大人ではありません。")

humans_data = {1: ("侍一郎", 21), 2:("侍次郎", 19), 3: ("侍三郎", 23)}

for person_id, (name, age) in humans_data.items():
  human_instance = Human(name, age)
  human_instance.check_adult()
```
